## O que é Bitcoin e outras moedas dessas na prática:
### Quando você compra uma moeda virtual e espera lucrar com isso é o mesmo que dizer que você espera que outras pessoas tbm comprem moeda virtual, pois só assim o valor delas é capaz de subir. Perceba: uma moeda virtual não é um ativo tal que seu valor se dá por algo tangível, ou seja, seu valor não se dá por um produto ou serviço valorizado e demandado pela sociedade, a partir do qual ela obtém receita e lucro, como é o caso das empresas. O valor desses ativos se dá pela demanda por ele. Mas note que a demanda por algo só pode existir caso as pessoas tenham motivos para desejarem esse algo. Dessa forma esse "algo" precisa proporcionar algum tipo de satisfação às pessoas. No caso de uma empresa lucrativa, as pessoas querem o papel dela porque ela proporciona retono real para elas, graças ao fato de que o produto ou serviço por ela ofertada à sociedade é desejada e útil a mesma e consequentemente ela consegue obter lucro a partir da oferta dessa bem ou serviço.

### O motivo para as pessoas demandarem criptoativos, dizem, ocorre pelo fato de que as pessoas querem fugir da "moeda estatal", mas isso é falso. As pessoas não querem isso. As pessoas compram Bitcoin, por exemplo, porque acham que o preço dela deve continuar subindo, ou seja, como já dito, que mais pessoas devem comprar Bitcoin ao longo do tempo levando a isso, e não porque querem fugir da inflação das moedas estatais e sei lá das quantas. Portanto, comprar criptoativo significa que você acredita que as pessoas vão continuar comprando disso, pois seu valor depende disso. No caso da Petrobrás: caso as pessoas deixem de comprar, o preço dela decai. Porém, caso ela obtenha lucro, as pessoas têm motivos reais para voltar a comprar dela, e comprarão. Ou seja, uma está vinculada à expectativa (moedas virtuais) e outra ao resultado real, mensurável e tangível.

### Vou dar o exemplo de uma empresa. No final das contas, o valor de mercado da Petrobrás vai sempre depender do lucro que ela obtém prestando seus serviços. Então, não importa o qe digam e nem nada: se a Petrobrás continuar e continuar dando lucro terá demanda por ativos dessa empresa. No caso do Bitcoin e de outras moedas virtuais, seu valor não está vinculado a nada. Ele depende da crença de que mais pessoas comprarão essa moeda e que consequentemente seu valor vai continuar subindo. Logo, perceba isso: se as pessoas pararem de comprar cripativos, ele perde todo o valor! A Petrobrás e outras empresas produtivas não. Caso todos decidam vender as ações da Petrobrás, mas ela obtenha lucro de 25 bi dólar, todos os que venderam vão correr para comprar disso novamente. Portanto, seu valor está vinculado a algo real, tangível e medivel! O Bitcoin não. O Bitcoin não tem algo real que o faça aumentar ou não de valor, mas apenas o palpite de que mais pessoas continurão comprando Bitcoin para fugir das moedas estatais e só. Caso as pessoas continuem comprando: o preço do Bitcoin sobe; caso não: ele decai.

## **Vamos ao código:**

## Carrego as bibliotecas:

```python
import yfinance as yf
import datetime as dt
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import seaborn as sns
```

## Lista de criptomoedas que serão avaliados:

```python
#lista de criptomoedas que buscarei os dados:
criptomoedas = {
    'Bitcoin': 'BTC-USD',
    'Ethereum': 'ETH-USD',
    'Tether': 'USDT-USD',
    'Binance Coin': 'BNB-USD',
    'USD Coin': 'USDC-USD',
    'XRP': 'XRP-USD',
    'Cardano': 'ADA-USD',
    'Dogecoin': 'DOGE-USD',
    'Polygon': 'MATIC-USD',
    'Solana': 'SOL-USD'
}
```

## Função para obter os dados de preços desses ativos:

```python
def dados_crypto():
    # Definir o intervalo de datas: de 01/01/2014 até 29/07/2024.
    start = dt.datetime(2019, 1, 3)
    end = dt.datetime(2024, 6, 29)

    # Criar um DataFrame vazio para armazenar os dados
    df = pd.DataFrame()

    # Baixar os dados de fechamento para cada criptomoeda e armazenar no DataFrame df
    for criptomoeda in criptomoedas.values():  # Aqui usamos .values() para acessar os tickers
        try:
            data = yf.download(criptomoeda, start=start, end=end)['Close']
            # Renomear a coluna do DataFrame para incluir o ticker
            df[criptomoeda.split('-')[0]] = data
        except Exception as e:
            print(f"Erro ao baixar dados para {criptomoeda}: {e}")

    # Resetar o índice para incluir a coluna de datas no DataFrame
    df.reset_index(inplace=True)

    return df  # Retorna o DataFrame criado
```

```python
# Chamada da função e atribuição do resultado a 'crypto'
cryptos = dados_crypto()

# Exibe as primeiras linhas do DataFrame
cryptos.head()
```

```python
#vou criar uma lista cores com a quantidade cores igual ao número de colunas do dataframe cryptos
cores = ['blue', 'red', 'darkgreen', 'orange', 'purple', 'pink', 'brown', 'gray', 'cyan', 'magenta']
```

## Plotar os dados simultaneamente:

```python
# Criar uma figura
plt.figure(figsize=(12, 6))

# Loop sobre as colunas do df_conservador para plotar os preços das ações
for i, coluna in enumerate(cryptos.columns[1:]):
    plt.plot(cryptos['Date'], cryptos[coluna], label=coluna, color = cores[i])

# Nomear os eixos do gráfico
plt.xlabel('Data')
plt.ylabel('Preço das Moedas')

# Adicionar título ao gráfico
plt.title('Preços das Moedas Virtuais ao Longo do Tempo')

# Adicionar legenda ao gráfico
plt.legend()

# Mostrar o gráfico
plt.show()
```

## Plotar os dados de preços de cada criptomoeda

```python
# Iterar sobre cada criptomoeda e plotar o gráfico correspondente
for i, crypto in enumerate(cryptos.columns[1:]):  # Começa do índice 1 para ignorar a coluna de datas
    plt.figure(figsize=(10, 6))
    plt.plot(cryptos['Date'], cryptos[crypto], label=crypto, color = cores[i])
    plt.title(f'Preço do {crypto}')
    plt.xlabel('Data')
    plt.ylabel('Preço em Dólar (USD)')

    #Colocar legenda com as labels configuradas
    plt.legend()

    #Mostrar o gráfico
    plt.show()
```

## Calculando a volatilidade:

```python
#Calcula o desvio padrão das moedas e armazena em uma variável (dp)
dp = cryptos.std()
```

```python
#Armazeno os dados de desvio padrão no dataframe, desvio_padrão
desvio_padrao = pd.DataFrame(dp)

#Coloco a coluna índice como fazendo parte do dataframe como coluna
desvio_padrao.reset_index(inplace = True)

#Apago a primiera coluna do dataframe, pq não vou precisar dela
desvio_padrão = desvio_padrao.drop(index = 0)

#Renomeio as colunas do dataframe
desvio_padrao.columns = ['Moeda', 'Dp']

#Exibo o dataframe
desvio_padrao
```

```python
# quero criar um loop para calcular o desvio padrão para todas as moedas e colocar em um gráfico
for i, crypto in enumerate(cryptos.columns[1:]):
    cryptos['dp_' + crypto] = cryptos[crypto].rolling(window=50).std()
    plt.figure(figsize=(10, 6))
    plt.plot(cryptos['Date'], cryptos['dp_' + crypto], label=crypto, color = cores[i])
    plt.title(f'Volatilidade do {crypto}')
    plt.xlabel('Data')
    plt.ylabel('Desvio Padrão')
    plt.legend()
    plt.show()
```

## A volatilidade desse tipo de ativo é gigantesca. As moedas que eu coloquei como "as que deram certo", ou seja, Bitcoin, Ethereum, Binance Coin e Solana têm isso como característica intrínseca.

# **Comparando a volatilidade do Bitcoin com o Ibovespa**

```python
# Baixar os dados do Ibovespa
ibovespa = yf.download('^BVSP', start='2019-01-03', end='2024-06-19')['Close']

# Colocar o índice como coluna
ibovespa_df = ibovespa.reset_index()

# Calcular a volatilidade utilizando uma janela móvel de 50 dias
ibovespa_df['volatility'] = ibovespa_df['Close'].rolling(window=50).std()

# Plotar os dados com Seaborn
plt.figure(figsize=(14, 7))
sns.lineplot(data=ibovespa_df, x='Date', y='volatility', color = 'green')
plt.title('Volatilidade do Ibovespa (Jan 2019 - Jun 2024)')
plt.xlabel('Data')
plt.ylabel('Volatilidade')
plt.show()
```

```python
#vou calcular a média da volatility
media_ibovespa = ibovespa_df['volatility'].mean()
print(f'A volatilidade média do Ibovespa é de: {media_ibovespa:.2f} pontos')
```

```python
#Escreve a última cotação do Bitcoin no período considerado
lucro_bitcoin = (cryptos['BTC'].iloc[-1] - cryptos['BTC'].iloc[0])/cryptos['BTC'].iloc[0] * 100
print(f"O lucro do Bitcoin é de {lucro_bitcoin:.2f}%")
valor_bit = (lucro_bitcoin / 100) * 1000
print(f'Se tivesse comprado R$ 1000 de Bitcoin em 2019, em Junho de 2024, teria, aproximadamente:')
print(f'R${valor_bit:.2f}!')
```

```python
for crypto in cryptos.columns[1:]:
    first_valid_index = cryptos[crypto].first_valid_index()
    if first_valid_index is not None:
        valor_compra = cryptos[crypto].iloc[first_valid_index]
        valor_final = cryptos[crypto].iloc[-1]  # Último valor na coluna, que corresponde à data final
        print(f"Para {crypto}:")
        print(f"Valor de compra: {valor_compra}")
        print(f"Valor final: {valor_final}")
        print(f"Diferença: {valor_final - valor_compra}")
        lucro = (valor_final - valor_compra) / valor_compra * 100
        print(f"Lucro: {lucro:.2f}%")
        percent = (lucro / 100)
        valor = percent * 1000
        valor = (lucro / 100) * 1000
        valor = round(valor, 2)
        print(f'Se vc tivesse comprado R$ 1000 de {crypto} hoje teria {valor}')

        print()
    else:
        print(f"Não foi possível encontrar um valor de compra válido para {crypto}.")
        print()
```
