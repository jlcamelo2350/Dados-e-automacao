### Defino as datas e crio as listas contendo as empresas que analisarei

```python
# Definir o intervalo de datas: de 01/01/2017 at√© 27/04/2024.
start = dt.datetime(2017, 1, 1)
end = dt.datetime(2024, 6, 10 )

# Lista de tickers das empresas

#Portf√≥lio conservador:
tickers_c = [
    'ITUB4.SA',  # Ita√∫ Unibanco
    'ABEV3.SA',  # Ambev
    'VALE3.SA',  # Vale
    'WEGE3.SA',  # Weg
    'BRFS3.SA',  # BRF
    'BBSE3.SA',  # Banco do Brasil Seguros
    'VBBR3.SA',  #Vibra Energia
    'UGPA3.SA'  # Ultrapar
             ]

#Portf√≥lio Equilibrado:
tickers_e = [
    'BBAS3.SA',  # Banco do Brasil
    'ITUB4.SA',  # Ita√∫ Unibanco
    'PETR4.SA',  # Petrobras
    'ITSA4.SA',  # Ita√∫sa
    'RAIZ4.SA',  # Ra√≠zen
    'RENT3.SA',  # Localiza
    'EMBR3.SA', # Embraer
    'CIEL3.SA',  # Cielo
    'SUZB3.SA',  # Suzano
    'JBSS3.SA'  #JBS
              ]

#Portf√≥lio mais Arrojado:
tickers_a = [
    'COGN3.SA',  # Cogna
    'CYRE3.SA',  # Cyrela
    'FLRY3.SA',  # Fleury
    'ROXO34.SA', # Nubank
    'PRIO3.SA',  # PetroRio
    'BRKM5.SA',   # Braskem
    'LREN3.SA',  # Lojas Renner
    'RAIL3.SA',  # Rumo SA
    'PETR4.SA',  # Petrobras
    'WEGE3.SA',  # Weg
    ]
```

### Obter os dados do portf√≥lio conservador:

```python
def dados_pre√ßos():

    # Criar um DataFrame vazio para armazenar os dados
    df = pd.DataFrame()

    # Baixar os dados de fechamento da empresa e armazenar no DataFrame df
    for ticker in tickers_c:
        try:
            data = yf.download(ticker, start=start, end=end)['Close']
            # Renomear a coluna do DataFrame para incluir o ticker
            df[ticker.split('.')[0]] = data
        except Exception as e:
            print(f"Erro ao baixar dados para {ticker}: {e}")

    # Resetar o √≠ndice para incluir a coluna de datas no DataFrame
    df.reset_index(inplace=True)

    return df
```

## **Calculando a variabilidade das empresas:**




#### O desvio padr√£o calculado a seguir indica o quanto os pre√ßos das a√ß√µes das empresas varia em rela√ß√£o √† m√©dia. Dito de outra forma, a partir desse c√°lculo simples conseguimos calcular o grau de variabilidade no valor desses ativos. Isso √© um bom indicador do risco do ativo em quest√£o, no entanto, perceba: apenas o c√°lculo do desvio padr√£o dos pre√ßos das a√ß√µes das empresas, n√£o fornece a informa√ß√£o completa acerca dos riscos dos pap√©is das mesmas.

### **Testando o desempenho das duas empresas com maior volatiidade:**
#### Vou colocar o desempenho da Vale e da WEG, em rela√ß√£o aos resultados delas √† prova em termos de Receitas e Lucros -- que no final √© o que importa para dizer se um investimento √© bom ou ruim:

###**Vale**

```python
def DRE():
    # Lista de tickers das empresas
    tickers = ['VALE3.SA']
    df1 = []

    # Loop para baixar a Demonstra√ß√£o de Resultados do Exerc√≠cio da empresa
    for ticker in tickers:
        ativo = yf.Ticker(ticker)
        dre = ativo.financials
        df1.append(dre.T)

    # Concatenar as Demonstra√ß√µes de Resultados do Exerc√≠cio em um √∫nico dataframe
    resultado = pd.concat(df1)
    resultado.reset_index(inplace=True)
    resultado.rename(columns={'index': 'Ano'}, inplace=True)
    return resultado
```

```python
# Chamar a fun√ß√£o DRE e exibir o resultado
dre_resultado = DRE()
dre_resultado.head()
```

```python
#Renomeando os nomes das colunas para Receita Total e Lucro L√≠quido:
dre_resultado.rename(columns={'Total Revenue': 'Receita Total', 'Net Income': 'Lucro L√≠quido'}, inplace=True)

#Reconfigurando as datas para mostrar apenas os anos:
dre_resultado['Ano'] = pd.to_datetime(dre_resultado['Ano']).dt.year
```

```python
# Criar uma figura:
plt.figure(figsize=(12, 6))

# Plotar os gr√°ficos:
plt.bar(dre_resultado['Ano'], dre_resultado['Receita Total'], label='Receita Total', color='green')
plt.scatter(dre_resultado['Ano'], dre_resultado['Receita Total'], color='black')
plt.plot(dre_resultado['Ano'], dre_resultado['Receita Total'], label = 'Tend√™ncia da Receita', color='black')
plt.bar(dre_resultado['Ano'], dre_resultado['Lucro L√≠quido'], label='Lucro L√≠quido', color='khaki')

# Nomear os eixos do gr√°fico
plt.xlabel('Ano')
plt.ylabel('Valor (em bilh√µes)')
plt.title('Receita Total e Lucro L√≠quido da Vale ao longo do tempo')

# Marcar os limites do eixo x
plt.xlim(2019, 2024)

# Mostrar a legenda
plt.legend()

# Mostrar as grades
plt.grid(True, alpha=0.2)

# Mostrar o gr√°fico
plt.show()
```

```python
print('Lucro da Vale ao longo do tempo:')

#Crio um dataframe lucro, contendo a coluna Lucro L√≠quido do dataframe, dre_resultado
lucro = pd.DataFrame(dre_resultado['Lucro L√≠quido'])

#Crio um dataframe per√≠odo, contendo a coluna Ano do dataframe, dre_resultado
per√≠odo = pd.DataFrame(dre_resultado['Ano'])

#Crio um dataframe lucro_periodo, contendo a conjun√ß√£o dos dataframes lucro e per√≠odo
lucro_periodo = pd.concat([per√≠odo, lucro], axis=1)

#Exibo o dataframe lucro_periodo na tela
print(lucro_periodo)
```

```python
def DRE_WEG():
    # Lista de tickers das empresas
    tickers = ['WEGE3.SA']
    df1 = []

    # Loop para baixar a Demonstra√ß√£o de Resultados do Exerc√≠cio da empresa
    for ticker in tickers:
        ativo = yf.Ticker(ticker)
        dre = ativo.financials
        df1.append(dre.T)

    # Concatenar as Demonstra√ß√µes de Resultados do Exerc√≠cio em um √∫nico dataframe
    resultado = pd.concat(df1)
    resultado.reset_index(inplace=True)
    resultado.rename(columns={'index': 'Ano'}, inplace=True)

    # Renomeando os nomes das colunas para Receita Total e Lucro L√≠quido
    resultado.rename(columns={'Total Revenue': 'Receita Total', 'Net Income': 'Lucro L√≠quido'}, inplace=True)

    # Reconfigurando as datas para mostrar apenas os anos
    resultado['Ano'] = pd.to_datetime(resultado['Ano']).dt.year

    return resultado
```
