#### Carrego as bibliotecas que utilizarei na análise

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

#### Defino as datas e crio as listas contendo as empresas que analisarei

```python
# Definir o intervalo de datas: de 01/01/2017 até 27/04/2024.
start = dt.datetime(2017, 1, 1)
end = dt.datetime(2024, 6, 10 )

# Lista de tickers das empresas

#Portfólio conservador:
tickers_c = [
    'ITUB4.SA',  # Itaú Unibanco
    'ABEV3.SA',  # Ambev
    'VALE3.SA',  # Vale
    'WEGE3.SA',  # Weg
    'BRFS3.SA',  # BRF
    'BBSE3.SA',  # Banco do Brasil Seguros
    'VBBR3.SA'   #Vibra Energia
             ]

#Portfólio Equilibrado:
tickers_e = [
    'BBAS3.SA',  # Banco do Brasil
    'PETR4.SA',  # Petrobras
    'ITSA4.SA',  # Itaúsa
    'RAIZ4.SA',  #Raízen
    'RENT3.SA',  # Localiza
    'EMBR3.SA',   # Embraer
    'SUZB3.SA'  #Suzano
              ]

#Portfólio mais Arrajado:
tickers_a = [
    'COGN3.SA',  # Cogna
    'CYRE3.SA',  # Cyrela
    'FLRY3.SA',  # Fleury
    'CIEL3.SA',  # Cielo
    'ROXO34.SA', # Nubank
    'PRIO3.SA',  # PetroRio
    'BRKM5.SA',   # Braskem
    'LREN3.SA',  # Lojas Renner
    ]
```

```python
def dados_preços():

    # Criar um DataFrame vazio para armazenar os dados
    df = pd.DataFrame()

    # Baixar os dados de fechamento da empresa e armazenar no DataFrame df
    for ticker in tickers_c:
        try:
            data = yf.download(ticker, start=start, end=end)['Close']
            # Renomear a coluna do DataFrame para incluir o ticker
            df[ticker.split('.')[0]] = data
        except Exception as e:
            print(f"Erro ao baixar dados para {ticker}: {e}")

    # Resetar o índice para incluir a coluna de datas no DataFrame
    df.reset_index(inplace=True)

    return df
```

```python
df_conservador = dados_preços()
```

### **Calculando a variabilidade das empresas:**

###**Vale**

```python
def DRE():
    # Lista de tickers das empresas
    tickers = ['VALE3.SA']
    df1 = []

    # Loop para baixar a Demonstração de Resultados do Exercício da empresa
    for ticker in tickers:
        ativo = yf.Ticker(ticker)
        dre = ativo.financials
        df1.append(dre.T)

    # Concatenar as Demonstrações de Resultados do Exercício em um único dataframe
    resultado = pd.concat(df1)
    resultado.reset_index(inplace=True)
    resultado.rename(columns={'index': 'Ano'}, inplace=True)
    return resultado
```

```python
# Chamar a função DRE e exibir o resultado
dre_resultado = DRE()
dre_resultado.head()
```

```python
#Renomeando os nomes das colunas para Receita Total e Lucro Líquido:
dre_resultado.rename(columns={'Total Revenue': 'Receita Total', 'Net Income': 'Lucro Líquido'}, inplace=True)

#Reconfigurando as datas para mostrar apenas os anos:
dre_resultado['Ano'] = pd.to_datetime(dre_resultado['Ano']).dt.year
```

```python
# Criar uma figura:
plt.figure(figsize=(12, 6))

# Plotar os gráficos:
plt.bar(dre_resultado['Ano'], dre_resultado['Receita Total'], label='Receita Total', color='green')
plt.scatter(dre_resultado['Ano'], dre_resultado['Receita Total'], color='black')
plt.plot(dre_resultado['Ano'], dre_resultado['Receita Total'], label = 'Tendência da Receita', color='black')
plt.bar(dre_resultado['Ano'], dre_resultado['Lucro Líquido'], label='Lucro Líquido', color='khaki')

# Nomear os eixos do gráfico
plt.xlabel('Ano')
plt.ylabel('Valor (em bilhões)')
plt.title('Receita Total e Lucro Líquido da Vale ao longo do tempo')

# Marcar os limites do eixo x
plt.xlim(2019, 2024)

# Mostrar a legenda
plt.legend()

# Mostrar as grades
plt.grid(True, alpha=0.2)

# Mostrar o gráfico
plt.show()
```

```python
print('Lucro da Vale ao longo do tempo:')

#Crio um dataframe lucro, contendo a coluna Lucro Líquido do dataframe, dre_resultado
lucro = pd.DataFrame(dre_resultado['Lucro Líquido'])

#Crio um dataframe período, contendo a coluna Ano do dataframe, dre_resultado
período = pd.DataFrame(dre_resultado['Ano'])

#Crio um dataframe lucro_periodo, contendo a conjunção dos dataframes lucro e período
lucro_periodo = pd.concat([período, lucro], axis=1)

#Exibo o dataframe lucro_periodo na tela
print(lucro_periodo)
```

```python
def DRE_WEG():
    # Lista de tickers das empresas
    tickers = ['WEGE3.SA']
    df1 = []

    # Loop para baixar a Demonstração de Resultados do Exercício da empresa
    for ticker in tickers:
        ativo = yf.Ticker(ticker)
        dre = ativo.financials
        df1.append(dre.T)

    # Concatenar as Demonstrações de Resultados do Exercício em um único dataframe
    resultado = pd.concat(df1)
    resultado.reset_index(inplace=True)
    resultado.rename(columns={'index': 'Ano'}, inplace=True)

    # Renomeando os nomes das colunas para Receita Total e Lucro Líquido
    resultado.rename(columns={'Total Revenue': 'Receita Total', 'Net Income': 'Lucro Líquido'}, inplace=True)

    # Reconfigurando as datas para mostrar apenas os anos
    resultado['Ano'] = pd.to_datetime(resultado['Ano']).dt.year

    return resultado
```
